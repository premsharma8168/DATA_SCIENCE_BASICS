# Airthmatic functions

```python
import numpy as np
var1 = np.array([91 , 12 , 13 , 4 , 5 , 6 , 7 , 8])
print("min : ",np.min(var1)) # min digit
print("_______________________________________________________________________________________________________")
print("max : ",np.max(var1)) # max digit
print("_______________________________________________________________________________________________________")
print("argmin : ",np.argmin(var1)) # min digit place
print("_______________________________________________________________________________________________________")
print("argmax : ",np.argmax(var1)) # max digit place
print("_______________________________________________________________________________________________________")
print("sqr root : ",np.sqrt(var1)) # 2 root
print("_______________________________________________________________________________________________________")
print("sin value : ",np.sin(var1)) # sin value on each
print("_______________________________________________________________________________________________________")
print("cos value : ",np.cos(var1)) # cos value on each
print("_______________________________________________________________________________________________________")
print("cumsum value : ",np.cumsum(var1)) # add before value and print like factorial is mul and it is add
```
