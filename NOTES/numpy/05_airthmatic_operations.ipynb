Addition

```python
import numpy as np
```

```python
a1 = np.array([1 , 2 , 3 , 4])
a1_add = a1 + 3
print(a1_add)
```

```python
# 1-D
var_a1 = np.array([1 , 2 , 3 , 4])
var_a2 = np.array([5 , 6 , 7 , 8])
varadd1 = np.add(var_a1 , var_a2)
print(varadd1)
```

```python
# 2-D
var_a11 = np.array([[1 , 2 , 3 , 4] , [5 , 6 , 7 , 8]])
var_a22 = np.array([[5 , 6 , 7 , 8] , [1 , 2 , 3 , 4]])
varadd2 = np.add(var_a11 , var_a22)
print(varadd2)
```

```python
# 3-D
var_a111 = np.array([[[1 , 2 , 3 , 4] , [5 , 6 , 7 , 8] , [5 , 6 , 7 , 8]]])
var_a222 = np.array([[[5 , 6 , 7 , 8] , [1 , 2 , 3 , 4] , [5 , 6 , 7 , 8]]])
varadd3 = np.add(var_a111 , var_a222)
print(varadd3)
```

Subtraction

```python
import numpy as np
```

```python
a1 = np.array([1 , 2 , 3 , 4])
a1_sub = a1 - 3
print(a1_sub)
```

```python
# 1-D
var_s1 = np.array([1 , 2 , 3 , 4])
var_s2 = np.array([5 , 6 , 7 , 8])
varsub1 = np.subtract(var_s1 , var_s2)
print(varsub1)
```

```python
# 2-D
var_s11 = np.array([[1 , 2 , 3 , 4] , [5 , 6 , 7 , 8]])
var_s22 = np.array([[5 , 6 , 7 , 8] , [1 , 2 , 3 , 4]])
varsub2 = np.subtract(var_s11 , var_s22)
print(varsub2)
```

```python
# 3-D
var_s111 = np.array([[[1 , 2 , 3 , 4] , [5 , 6 , 7 , 8] , [5 , 6 , 7 , 8]]])
var_s222 = np.array([[[5 , 6 , 7 , 8] , [1 , 2 , 3 , 4] , [5 , 6 , 7 , 8]]])
varsub1 = np.subtract(var_s111 , var_s222)
print(varsub1)
```
