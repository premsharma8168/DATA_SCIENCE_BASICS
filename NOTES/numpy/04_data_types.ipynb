# Data Types as a Function

_________________________________________________________________________________________

integer

```python
import numpy as np
```

```python
var1 = np.array([1 , 2 , 3 , 4 , 45 , 65 , 28 , 91 , 73])
print("Data Type : " , var1.dtype)
print(var1)
```

_________________________________________________________________________________________

float

```python
import numpy as np
```

```python
var2 = np.array([1.1 , 2.2 , 3.3])
print("Data Type : " , var2.dtype)
print(var2)
```

_________________________________________________________________________________________

string

```python
import numpy as np
```

```python
var3 = np.array(["p" , "r" , "e" , "m"])
print("Data Type : " , var3.dtype)
print(var3)
```

_________________________________________________________________________________________

string mix with int

_________________________________________________________________________________________

# changing type of one to another

to int8

```python
import numpy as np
```

```python
x = np.array([1 , 2 , 3 , 4 , 45 , 65 , 28 , 91 , 73] , dtype=np.int8)
print("Data Type : " , x.dtype)
print(x)
```

_____________________________________________________________________________________________

to float

```python
x1 = np.array([1 , 2 , 3 , 4 , 45 , 65 , 28 , 91 , 73] , dtype="f")
print("Data Type : " , x1.dtype)
print(x1)
```

-----------------------------------------------------------------------------------------

other method

```python
xx = np.array([1 , 2 , 3 , 4 , 45 , 65 , 28 , 91 , 73])

new = np.float32(xx)
newx = np.int_(new)
print("Data Type : " , xx.dtype)
print("Data Type : " , new.dtype)
print("Data Type : " , newx.dtype)
print(xx)
print(new)
print(newx)
```

-----------------------------------------------------------------------------------------

astype function

```python
xxx = np.array([1 , 2 , 3 , 4 , 45 , 65 , 28 , 91 , 73])

newxxx = xxx.astype(float)

print(xxx)
print(newxxx)
```
