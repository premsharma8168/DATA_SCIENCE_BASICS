```python
# Problem 1: Plotting a Linear Function
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = 2*x + 3
plt.figure()
plt.plot(x, y)
plt.title('Linear Function: y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Problem 2: Plotting Multiple Functions
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = np.sqrt(x)
plt.figure()
plt.plot(x, y1, label='y = x^2', color='blue')
plt.plot(x, y2, label='y = sqrt(x)', color='red', linestyle='--')
plt.title('Multiple Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Problem 3: Plotting Trigonometric Functions
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
plt.figure()
plt.plot(x, y_sin, label='y = sin(x)', color='blue')
plt.plot(x, y_cos, label='y = cos(x)', color='red')
plt.plot(x, y_tan, label='y = tan(x)', color='green')
plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Problem 4: Plotting Exponential Growth
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-5, 5, 100)
y = np.exp(x)
plt.figure()
plt.plot(x, y)
plt.title('Exponential Growth: y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Problem 5: Plotting Piecewise Functions
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-2, 2, 100)
y = np.where(x < 0, -1, 1)
plt.figure()
plt.plot(x, y)
plt.title('Piecewise Function: f(x) = -1 if x < 0, 1 if x >= 0')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Problem 6: Plotting Parametric Equations
from matplotlib import pyplot as plt
import numpy as np
p = np.linspace(0, 2*np.pi, 100)
x = np.cos(p)
y = np.sin(p)
plt.figure()
plt.plot(x, y)
plt.title('Parametric Equations: x = cos(t), y = sin(t)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Problem 7: Plotting Logarithmic Functions
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(0.1, 10, 100)
y1 = np.log(x)
y2 = np.log10(x)
plt.figure()
plt.plot(x, y1, label='y = log(x)', color='blue')
plt.plot(x, y2, label='y = log10(x)', color='red')
plt.title('Logarithmic Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Problem 8: Plotting Step Function
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-5, 5, 100)
y = np.where(x < 0, 0, 1)
plt.figure()
plt.plot(x, y)
plt.title('Step Function: f(x) = 0 if x < 0, 1 if x >= 0')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Problem 9: Plotting Polynomial Functions
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 100)
y = x**3 - 2*x**2 + x - 1
plt.figure()
plt.plot(x, y)
plt.title('Polynomial Function: y = x^3 - 2x^2 + x - 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Problem 10: Plotting Piecewise Linear Functions
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-2, 2, 100)
y = np.where(x <= 0, -x, x)
plt.figure()
plt.plot(x, y)
plt.title('Piecewise Linear Function: f(x) = -x if x <= 0, x if x > 0')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
